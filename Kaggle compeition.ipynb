```python
columns=['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7', 'cat8','cat9']
for x in columns:
    print(x)
    print(train[x].value_counts())
```

```python
for x in columns:
    print(x)
    print(test[x].value_counts())
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

train[columns]=ordinal_encoder.fit_transform(train[columns])
test[columns]=ordinal_encoder.transform(test[columns])
train.head()
```

```python
train.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train.loc[:,train.columns!='target'],train['target'], 
                test_size = 0.20, random_state = 0)
```

```python
X_train.head()
```

```python

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
lr=LinearRegression()
lr.fit(X_train,y_train)
predictions=lr.predict(X_test)
mae=mean_absolute_error(y_test,predictions)
print(mae)
r2=r2_score(y_test,predictions)
print(r2)
mse=mean_squared_error(y_test,predictions,squared=False)
print(mse)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
gbr.fit(X_train,y_train)
predictions=gbr.predict(X_test)
mae=mean_absolute_error(y_test,predictions)
print(mae)
r2=r2_score(y_test,predictions)
print(r2)
mse=mean_squared_error(y_test,predictions,squared=False)
print(mse)
```

```python
from xgboost import XGBRegressor
xgbr=XGBRegressor(learning_rate=0.1,n_estimators=200,gamma=0.1,use_label_encoder=False)
xgbr.fit(X_train, y_train, 
             early_stopping_rounds=10, 
             eval_set=[(X_test, y_test)], 
             verbose=False)
predictions=xgbr.predict(X_test)
mae=mean_absolute_error(y_test,predictions)
print(mae)
r2=r2_score(y_test,predictions)
print(r2)
mse=mean_squared_error(y_test,predictions,squared=False)
print(mse)
```

```python
from catboost import CatBoostRegressor
cbr = CatBoostRegressor(random_state=1,loss_function='RMSE',depth=2,learning_rate=0.5,l2_leaf_reg=8,iterations=1000)
cbr.fit(X_train, y_train,eval_set=[(X_test, y_test)],early_stopping_rounds=50,verbose=False)
predictions=cbr.predict(X_test)
mae=mean_absolute_error(y_test,predictions)
print(mae)
r2=r2_score(y_test,predictions)
print(r2)
mse=mean_squared_error(y_test,predictions,squared=False)
print(mse)
```

```python
from lightgbm import LGBMRegressor
model = LGBMRegressor()
model.fit(X_train,y_train)
predictions=model.predict(X_test)
mae=mean_absolute_error(y_test,predictions)
print(mae)
r2=r2_score(y_test,predictions)
print(r2)
mse=mean_squared_error(y_test,predictions,squared=False)
print(mse)
```

```python
new_prediction = cbr.predict(test)

# Save the predictions to a CSV file
output = pd.DataFrame({'Id': test.index,
                       'target': new_prediction})
output.to_csv('submission.csv', index=False)
```
